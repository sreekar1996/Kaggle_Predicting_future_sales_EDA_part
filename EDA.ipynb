```python
from datetime import datetime
```

```python
def getmonth(date):
	date = datetime.strptime(date, '%d.%m.%Y')
	return date.month
```

```python

getmonth('01.10.2015')
```

```python
day = []
for d in df1.date.values:
	day.append(getmonth(d))
```
